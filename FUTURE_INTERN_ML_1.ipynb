```python
# Install if needed
# !pip install prophet pandas matplotlib

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from prophet import Prophet

# Step 1: Load data
# Sample simulated data - replace this with your real dataset
# Format required: Date column -> "ds", Sales column -> "y"

# Simulating monthly sales data
date_rng = pd.date_range(start='2018-01-01', end='2023-12-01', freq='MS')
sales_data = pd.DataFrame(date_rng, columns=['ds'])
sales_data['y'] = 5000 + (sales_data.index * 50) + (1000 * np.sin(2 * np.pi * sales_data.index / 12)) + np.random.normal(0, 200, len(sales_data))

# Step 2: Initialize and fit Prophet model
model = Prophet(yearly_seasonality=True)
model.fit(sales_data)

# Step 3: Make future dataframe and forecast
future = model.make_future_dataframe(periods=12, freq='MS')  # Forecasting next 12 months
forecast = model.predict(future)

# Step 4: Visualize forecast
fig1 = model.plot(forecast)
plt.title("Sales Forecast")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.show()

# Step 5: Visualize trend and seasonality
fig2 = model.plot_components(forecast)
plt.show()

# Step 6: Export forecast data for Power BI
forecast_output = forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]
forecast_output.to_csv('/content/Data.csv', index=False)
```

```python
# Install if needed
# !pip install prophet pandas matplotlib

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from prophet import Prophet

# Helper function to build and forecast
def run_forecast(date_rng, freq, granularity, trend_coef, seasonality_amp, noise_std, periods):
    df = pd.DataFrame({'ds': date_rng})
    df['y'] = (
        1000 +
        np.arange(len(date_rng)) * trend_coef +
        seasonality_amp * np.sin(2 * np.pi * np.arange(len(date_rng)) / len(date_rng)) +
        np.random.normal(0, noise_std, len(date_rng))
    )

    # Initialize Prophet model
    if freq == 'D':
        model = Prophet(daily_seasonality=True)
    elif freq.startswith('W'):
        model = Prophet(weekly_seasonality=True)
    else:
        model = Prophet(yearly_seasonality=True)

    model.fit(df)

    # Forecast
    future = model.make_future_dataframe(periods=periods, freq=freq)
    forecast = model.predict(future)

    # Save plot (optional)
    model.plot(forecast)
    plt.title(f"{granularity} Sales Forecast")
    plt.xlabel("Date")
    plt.ylabel("Sales")
    plt.show()

    model.plot_components(forecast)
    plt.show()


    forecast = forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]
    forecast['granularity'] = granularity

    return forecast

# 1. Monthly
monthly_dates = pd.date_range(start='2018-01-01', end='2023-12-01', freq='MS')
monthly_forecast = run_forecast(
    date_rng=monthly_dates,
    freq='MS',
    granularity='Monthly',
    trend_coef=50,
    seasonality_amp=1000,
    noise_std=200,
    periods=12
)

# 2. Weekly
weekly_dates = pd.date_range(start='2020-01-05', end='2023-12-31', freq='W-SUN')
weekly_forecast = run_forecast(
    date_rng=weekly_dates,
    freq='W',
    granularity='Weekly',
    trend_coef=10,
    seasonality_amp=600,
    noise_std=150,
    periods=12
)

# 3. Daily
daily_dates = pd.date_range(start='2023-01-01', end='2023-12-31', freq='D')
daily_forecast = run_forecast(
    date_rng=daily_dates,
    freq='D',
    granularity='Daily',
    trend_coef=0.5,
    seasonality_amp=100,
    noise_std=30,
    periods=30
)

# Export each separately
monthly_forecast.to_csv("/content/monthly_forecast.csv", index=False)
weekly_forecast.to_csv("/content/weekly_forecast.csv", index=False)
daily_forecast.to_csv("/content/daily_forecast.csv", index=False)

# Combine all into one file for Power BI (optional)
combined = pd.concat([monthly_forecast, weekly_forecast, daily_forecast], ignore_index=True)
combined.to_csv("/content/all_granularities_forecast.csv", index=False)

print("Forecasts saved for all granularities.")
```
